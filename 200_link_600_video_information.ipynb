```python
import numpy as np
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
driver = webdriver.Chrome(ChromeDriverManager().install())
#driver.get("https://www.youtube.com/")
time.sleep(5)
```

```python
link = pd.read_csv("200_link.csv")
list_link=list(link["Link"])
type(list_link)
```

```python
list_link
```

```python
def listToString(s): 
    str1 = " " 
    
    # return string  
    return (str1.join(s))

def get_duration(s,name):
    #s=s.replace(" ","")
    s=s.replace(" \n ","")
    s=s.replace("\n","")
    s=s.replace("Skip ","")
    s=s.replace("navigation ","")
    s=s.replace("Sign in ","")
    s=s.replace("Subscribe ","")
    s=s.replace("Play all ","")
    s=s.replace("Join ","")
    s=s.replace("Membership ","")
    s=s.replace("Got It ","")
    s=s.replace(name,"")
    return s
```

```python
data=[]
for i in list_link:
    duration=[]
    title=[]
    link=[]
    view_no=[]
    descriptions=[]
    commands=[]
    likes=[]
    dislikes=[]

    video_link=i+"/videos"
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(video_link)
    time.sleep(3)


    title_element=driver.find_elements_by_xpath('//*[@id="video-title"]')
    duration_element=driver.find_elements_by_xpath('//*[@id="text"]')
    view_element=driver.find_elements_by_xpath('//*[@id="metadata-line"]/span[1]')

    dur=[]
    for i in duration_element:
        dur.append(i.get_attribute('textContent'))
    name=dur[2]


    s=listToString(dur)

    sa=get_duration(s,name)
    sa=sa.split("  ")
    
    duration.append(sa[0])
    if len(sa)>1:
        duration.append(sa[1])
    if len(sa)>2:
        duration.append(sa[2])

    for i in view_element[:3]:
        view_no.append(i.get_attribute('textContent'))

    for i in title_element[:3]:
        title.append(i.get_attribute('textContent'))
        l=i.get_attribute('href')
        link.append(l)
        vid = webdriver.Chrome(ChromeDriverManager().install())
        vid.get(l)
        time.sleep(3)
        dis=vid.find_elements_by_xpath('//*[@class="style-scope yt-formatted-string"]')
        d=[]
        for i in dis:
            d.append(i.get_attribute('textContent'))
        description=listToString(d)
        description=description.replace("\n","")
        descriptions.append(description)

        like=[]
        duration_element=vid.find_elements_by_xpath('//*[@id="text"]')


        for i in duration_element:
            like.append(i.get_attribute('textContent'))

        lik=like[2]
        dislik=like[3]
        likes.append(lik)
        dislikes.append(dislik)
        while True:
            com=vid.find_elements_by_xpath('//*[@class="count-text style-scope ytd-comments-header-renderer"]/span[1]')
            scroll_height = 2000
            h1 = vid.execute_script("return document.documentElement.scrollHeight")
            vid.execute_script(f"window.scrollTo(0, {h1 + scroll_height});")
            time.sleep(1.5)
            h2 = vid.execute_script("return document.documentElement.scrollHeight")
            if h1 == h2:
                break 
        for i in com:
            command=i.get_attribute('textContent') 

        commands.append(command)
        vid.quit()


    driver.quit()
    data.append([video_link,"video-1",title[0],duration[0],link[0],view_no[0],descriptions[0],commands[0],likes[0],dislikes[0]])
    if len(title)>2:
        data.append([video_link,"video-2",title[1],duration[1],link[1],view_no[1],descriptions[1],commands[1],likes[1],dislikes[1]])
    if len(title)>2:
        data.append([video_link,"video-3",title[2],duration[2],link[2],view_no[2],descriptions[2],commands[2],likes[2],dislikes[2]])
```

```python
column=["Channel_link","Video-No","Video Title","Duration","Video link","view_no","Descriptions","No of Commands","No of  likes","No of dislikes"]
df = pd.DataFrame(data,columns=column)
df
df.to_csv('200_link_600_videos_data_scrabing_data.csv')
```
